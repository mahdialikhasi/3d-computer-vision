# Problem 1 : Haris corner detector

### part a : compute x and y derivatives

```python
def derivatives(image):
    sobelx = cv2.Sobel(image,cv2.CV_64F,1,0,ksize=3)
    sobely = cv2.Sobel(image,cv2.CV_64F,0,1,ksize=3)
    return (sobelx, sobely)
```

### part b : covariance matrix H

```python
def gauss_kernel(size = 3,sigma = 1):
    # create a gaussian kernel using numpy
    center = int(size/2)
    kernel = np.zeros((size,size))
    
    for i in range(size):
        for j in range(size):
            diff = np.sqrt((i - center) ** 2 + (j - center) ** 2)
            kernel[i,j] = np.exp(-(diff ** 2) / (2 * sigma ** 2))
    return kernel / np.sum(kernel)

def covariance_image(Ix, Iy, window_size):
    if(window_size % 2 != 1):
        window_size += 1
        
    output = np.zeros((Ix.shape[0], Ix.shape[1], 4), 'double')
    
    # pad array
    Ix_p = np.pad(Ix, (int(window_size / 2), int(window_size / 2)))
    Iy_p = np.pad(Iy, (int(window_size / 2), int(window_size / 2)))
    
    # gaussian matrix
    kernel = gauss_kernel(window_size)
    
    for i in range(len(Ix_p) - window_size + 1):
        for j in range(len(Ix_p) - window_size + 1):
            windowed_Ix = Ix_p[i:i+window_size,j:j+window_size]
            windowed_Iy = Iy_p[i:i+window_size,j:j+window_size]
            x2 = windowed_Ix * windowed_Ix
            y2 = windowed_Iy * windowed_Iy
            xy = windowed_Ix * windowed_Iy
            # calculate covariance matrix of this window
            H = np.array([np.sum(x2 * kernel), np.sum(xy * kernel), np.sum(xy * kernel), np.sum(y2 * kernel)])
            output[i,j] = H
            
    return output
```

### part c : compute Haris response

```python
def haris_response(H):
    output = np.zeros((H.shape[0], H.shape[1]), 'double')
    for i in range(H.shape[0]):
        for j in range(H.shape[1]):
            if H[i,j,0] + H[i,j,3] != 0 :
                output[i, j] = (H[i,j,0] * H[i,j,3] - H[i,j,1] * H[i,j,2]) / (H[i,j,0] + H[i,j,3])
    return output
```

### part d : Find peaks in the response that are above the threshold

```python
def haris_peaks(image, treshold):
    return np.where(image >= treshold)
```

### part e : run on boxes image

```python
# import image
image = cv2.imread('images/Boxes.png', 0)
Ix, Iy = derivatives(image)
H = covariance_image(Ix, Iy, 9)
alpha = haris_response(H)
```

```python
interest_points_x, interest_points_y = haris_peaks(alpha, 100000)

# show the plot
plt.imshow(image, cmap='gray')
plt.plot(interest_points_y, interest_points_x, 'yo')
```

# Problem Two : Keypoint Matching

### part a : Compute the descriptors for each interest point

```python
def descriptor(image, keypoints):
    # Initiate STAR detector
    orb = cv2.ORB_create()
    
    # compute the descriptors with ORB
    kp, des = orb.compute(image, keypoints)
    
    return (kp, des)

def descriptor(image):
    # Initiate STAR detector
    sift = cv2.SIFT_create()
    
    # find the keypoints with ORB
    kp = sift.detect(image,None)
    
    # compute the descriptors with ORB
    kp, des = sift.compute(image, kp)
    
    return (kp, des)
```

### part b, c : For each interest point in image 1, find its best match in image 2. Add the pair of matching points to the list of matches

```python
def ssd(arr1, arr2):
    x = np.array(arr1 - arr2)
    return np.matmul(x, x.transpose())
    
def matching(kp1, kp2, des1, des2, max_matching = 500):
    match = []
    distance = np.zeros(len(kp1), 'double')
    match_array = []
    for i in range(len(kp1)):
        minimum = math.inf
        index = 0
        for j in range(len(kp2)):
            diff = ssd(des1[i], des2[j])
            if(minimum > diff):
                minimum = diff
                index = j
                distance[i] = diff
        match.append(kp2[index])
        match_array.append(cv2.DMatch(i, j, distance[i]))
    
    kp1 = np.array(kp1)
    match = np.array(match)
    match_array = np.array(match_array)
    if(len(kp1) < max_matching):
        return (kp1, match, match_array)
    else:
        index_list = sorted(range(len(distance)), key=lambda i: distance[i])[:max_matching]
        return (kp1[index_list], match[index_list], match_array[index_list])
```

### part d : Display the matches

```python
def pk2point(pk):
    output = np.zeros((len(pk), 2))
    for i in range(len(pk)):
        output[i] = pk[i].pt
    return output

def compute_fundamental(points1, points2):
    return cv2.findFundamentalMat(points1, points2, cv2.FM_8POINT)
    
def display(image2, points1, points2, f_matrix):
    plt.axis([0, 530, 400, 0])
    plt.imshow(image2)
    plt.plot(points2[:,0], points2[:,1], 'yo')
    for i in range(len(points1)):
        t = np.matmul(f_matrix[0], np.array([np.append(points1[i], [1])]).transpose())[:,0]
        x = np.linspace(0,600,100)
        y = (-t[2] -x * t[0])/ t[1]
        plt.plot(x, y, '-r')
        
def display_mathes(img1,kp1,img2,kp2,matches):
    return cv2.drawMatchesKnn(img1, kp1, img2, kp2, matches1to2 = matches, outImg = None, flags=2)
```

### part e : Compute the Harris corner detector and find matches for images ”Rainier1.png” and ”Rainier2.png”.

```python
# find points and their descriptor
im1_points, im1_desc = descriptor(image1_gray)
im2_points, im2_desc = descriptor(image2_gray)

points_match1, points_match2, match_array = matching(im1_points, im2_points, im1_desc, im2_desc, 100)
points1 = pk2point(points_match1)
points2 = pk2point(points_match2)
f = compute_fundamental(points1, points2)

display(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB), points1, points2, f)
```

### part f : using RANSAK

```python
def compute_fundamental_RANSAC(points1, points2, n = 1000):
    # return cv2.findFundamentalMat(points1, points2, cv2.FM_RANSAC, ransacReprojThreshold = 1, confidence = 0.99)
    # compute fundamental matrix using 8-algorithm and ransac
    max_inliers = 0
    for i in range(n):
        # select 8 random point
        a = list(range(len(points1)))
        random.shuffle(a)
        subgroup1 = points1[a[0:8]]
        subgroup2 = points2[a[0:8]]

        # calculate fundamental matrix
        f_matrix = cv2.findFundamentalMat(subgroup1, subgroup2, cv2.FM_8POINT)
        if f_matrix[0] is None:
            continue
        
        # calculate inliers
        inliers_count = 0
        for j in range(len(points1)):
            d1 = np.matmul(np.array([np.append(points2[j], [1])]), 
                         np.matmul(f_matrix[0], np.array([np.append(points1[j], [1])]).transpose())
                         )
            d2 = np.matmul(f_matrix[0], np.array([np.append(points1[j], [1])]).transpose())
            d3 = np.matmul(np.array([np.append(points2[j], [1])]), f_matrix[0])
            distance = (d1[0,0] ** 2) * (
                1 / (d2[0,0] ** 2 + d2[1,0] ** 2) + 1/ (d3[0,0] ** 2 + d3[0,1] ** 2))
            if(distance <= 3):
                inliers_count += 1
        if(inliers_count > max_inliers):
            max_inliers = inliers_count
            best_f = f_matrix
            best_index = a[0:8]

    return best_f, best_index
```

```python
# find points and their descriptor
im1_points, im1_desc = descriptor(image1_gray)
im2_points, im2_desc = descriptor(image2_gray)

points_match1, points_match2, match_array = matching(im1_points, im2_points, im1_desc, im2_desc, 100)
points1 = pk2point(points_match1)
points2 = pk2point(points_match2)
f, best_index = compute_fundamental_RANSAC(points1, points2)

display(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB), points1, points2, f)
```

# Problem Three : Image Stitching

### part a : Compute the size of stitchedImage.

```python
def stiched_image_size(image1, image2, h):
    h_inv = np.linalg.inv(h)
    points = np.array([
        [0,0],
        [image2.shape[1], 0],
        [0, image2.shape[0]],
        [image2.shape[1], image2.shape[0]]
    ], dtype=np.float32)
    points = np.array([points])
    new =  cv2.perspectiveTransform(points, h_inv)
    new = new[0]
    # return min_x, max_x, min_y, max_y, stiched_image
    min_x = np.min(new[:,0])
    max_x = np.max(new[:,0])
    min_y = np.min(new[:,1])
    max_y = np.max(new[:,1])
                   
    return min_x, max_x, min_y, max_y, np.zeros(
        (
            math.ceil(max(max_y, image1.shape[0]) - min(0,min_y)),
            math.ceil(max(max_x, image1.shape[1]) - min(0,min_x)),
            3
        ),
        dtype="int"
    )
    
```

### part b : Copy ”image1” onto the ”stitchedImage” at the right location

```python
def copy_image(stiched_image, image, min_x, min_y):
    min_x = min(math.ceil(min_x), 0)
    min_y = min(math.ceil(min_y), 0)
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            stiched_image[i - min_y, j - min_x] = image[i,j]
    return stiched_image
```

### part c : For each pixel in ”stitchedImage”, project the point onto ”image2”

```python
def project_image(stiched_image, min_x, min_y, image2, h):
    for i in range(stiched_image.shape[0]):
        for j in range(stiched_image.shape[1]):
            point = np.array([[j + min(min_x,0), i + min(min_y,0)]])
            point = np.array([point], dtype = np.float64)
            im2_point = cv2.perspectiveTransform(point, h)
            im2_point = im2_point[0][0]
            if(im2_point[0] > 0 and math.ceil(im2_point[0] + 1) < image2.shape[1] and 
               im2_point[1] > 0 and math.ceil(im2_point[1] + 1) < image2.shape[0]) : 
                    # bilinear interpolation
                    wx = (im2_point[0] - math.floor(im2_point[0])) / (math.ceil(im2_point[0] + 1) - math.floor(im2_point[0]))
                    wy = (im2_point[1] - math.floor(im2_point[1])) / (math.ceil(im2_point[1] + 1) - math.floor(im2_point[1]))
                    A = image2[math.floor(im2_point[1]), math.floor(im2_point[0])]
                    B = image2[math.floor(im2_point[1]), math.ceil(im2_point[0] + 1)]
                    c = image2[math.ceil(im2_point[1] + 1), math.floor(im2_point[0])]
                    D = image2[math.ceil(im2_point[1] + 1), math.ceil(im2_point[0] + 1)]
                    stiched_image[i,j] = A*(1-wx)*(1-wy)+B*wx*(1-wy)+c * (1 - wx) * (wy) + D * (wx) * (wy)
    return stiched_image
```

### part d : mosaicing

```python
# import image
image1 = cv2.imread('images/Rainier1.png')
image1_gray = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
image1_rgb = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)

image2 = cv2.imread('images/Rainier2.png')
image2_gray = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
image2_rgb = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
```

```python
def compute_h(points1, points2):
    return cv2.findHomography(points1, points2, cv2.RANSAC, ransacReprojThreshold = 5, confidence = 0.99)
```

```python
points_match1, points_match2, match_array = matching(im1_points, im2_points, im1_desc, im2_desc, 100)
points1 = pk2point(points_match1)
points2 = pk2point(points_match2)
h = compute_h(points1, points2)
```

```python
min_x, max_x, min_y, max_y, stitched_image = stiched_image_size(image1_rgb, image2_rgb, np.array(h[0]))
stitched_image = copy_image(stitched_image, image1_rgb, min_x, min_y)
stitched_image = project_image(stitched_image, min_x, min_y, image2_rgb, np.array(h[0]))
plt.imshow(stitched_image)
```

# Problem Four : Panorama

### Part a : Create a panorama that stitches together the six Mt. Rainier photographs

```python
# import image
image1 = cv2.imread('images/Rainier1.png')
image1_gray = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
image1_rgb = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)

image2 = cv2.imread('images/Rainier2.png')
image2_gray = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
image2_rgb = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB) 

# find points and their descriptor
im1_points, im1_desc = descriptor(image1_gray)
im2_points, im2_desc = descriptor(image2_gray)
```

```python
points_match1, points_match2, match_array = matching(im1_points, im2_points, im1_desc, im2_desc, 100)
points1 = pk2point(points_match1)
points2 = pk2point(points_match2)
h = compute_h(points1, points2)
```

```python
min_x, max_x, min_y, max_y, stitched_image = stiched_image_size(image1_rgb, image2_rgb, np.array(h[0]))
all_stitched_image = copy_image(stitched_image, image1_rgb, min_x, min_y)
all_stitched_image = project_image(stitched_image, min_x, min_y, image2_rgb, np.array(h[0]))
```

```python
for i in range(3, 7):
    image = cv2.imread('images/Rainier' + str(i) + '.png')
    image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # find points and their descriptor
    im1_points, im1_desc = descriptor(cv2.cvtColor(np.array(np.uint8(all_stitched_image)), cv2.COLOR_RGB2GRAY))
    im2_points, im2_desc = descriptor(image_gray)
    
    # find matches and h
    points_match1, points_match2, match_array = matching(im1_points, im2_points, im1_desc, im2_desc, 100)
    points1 = pk2point(points_match1)
    points2 = pk2point(points_match2)
    h = compute_h(points1, points2)
    
    min_x, max_x, min_y, max_y, all_stitched_tmp = stiched_image_size(
        np.array(np.uint8(all_stitched_image)), image_rgb, np.array(h[0]))
    all_stitched_tmp = copy_image(all_stitched_tmp, np.array(np.uint8(all_stitched_image)), min_x, min_y)
    all_stitched_tmp = project_image(all_stitched_tmp, min_x, min_y, image_rgb, np.array(h[0]))
    all_stitched_image = all_stitched_tmp
```
